## Description
You are given a table, BST, containing two columns: N and P, where N represents the value of a node in Binary Tree, and P is the parent of N.

![Table](https://s3.amazonaws.com/hr-challenge-images/12888/1443818507-5095ab9853-1.png)

## To Do
Write a query to find the node type of Binary Tree ordered by the value of the node. Output one of the following for each node:

    Root: If node is root node.
    Leaf: If node is leaf node.
    Inner: If node is neither root nor leaf node.

## Importing

````python
import pandas as pd, numpy as np
from pandasql import sqldf
````

## Define Schema

````python
# Define the data as a list of lists
data = {
    'N': [1, 3, 5, 7, 2, 6, 4, 8, 10, 12, 14, 9, 13, 11, 15],
    'P': [2, 2, 6, 6, 4, 4, 15, 9, 9, 13, 13, 11, 11, 15, np.nan]
}

df = pd.DataFrame(data)

df
````

````python
# Define the data as a list of lists
data = {
    'N': [1, 3, 6, 9, 2, 8, 5, ],
    'P': [2, 2, 8, 8, 5, 5, np.nan]
}

df_2 = pd.DataFrame(data)

df_2
````

## Task

````python
# Define the SQL query
query = """
WITH tree_family AS (
    SELECT DISTINCT b1.N AS N, b1.P AS P, b2.P AS PT
    FROM BST AS b1
    LEFT JOIN BST AS b2 On b1.N = b2.P 
)
SELECT 
    N, 
    CASE 
        WHEN P IS NULL THEN 'Root'
        WHEN P IS NOT NULL AND PT IS NULL THEN 'Leaf'
        ELSE 'Inner'
    END AS type
FROM tree_family
ORDER BY N
;
"""

# Excute the query using pandasql
result = sqldf(query, env={'BST':df})

# Display the result dataframe
display(result)
````

### Another Solution

````python
# Define the SQL query
query = """
SELECT N,
    CASE
        WHEN P IS NULL THEN 'Root'
        WHEN N IN (SELECT P FROM BST) THEN 'Inner'
        ELSE 'Leaf'
    END
FROM BST
ORDER BY N ASC
;
"""

# Excute the query using pandasql
result = sqldf(query, env={'BST':df})

# Display the result dataframe
display(result)
````
