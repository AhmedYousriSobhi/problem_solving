## Importing

```python
import pandas as pd
from pandasql import sqldf
```

## Define Schema

```python
# Create a list of tuples with the data
data = [
    [1, "Ashley", 2340],
    [2, "Julia", 1198],
    [3, "Britney", 9009],
    [4, "Kristeen", 2341],
    [5, "Dyana", 9990],
    [6, "Diana", 8011],
    [7, "Jenny", 2341],
    [8, "Christeen", 2342],
    [9, "Meera", 2343],
    [10, "Priya", 2344],
    [11, "Priyanka", 2345],
    [12, "Paige", 2346],
    [13, "Jane", 2347],
    [14, "Belvet", 2348],
    [15, "Scarlet", 2349],
    [16, "Salma", 9087],
    [17, "Amanda", 7777],
    [18, "Aamina", 5500],
    [19, "Amina", 2570],
    [20, "Ketty", 2007]
]

columns = ['ID', 'NAME', 'Salary']

df = pd.DataFrame(data, columns=columns)

df
```

## Task

```python
# Define the SQL query
query = """
SELECT CEIL(AVG(Salary) - AVG(REPLACE(Salary, '0', '')))
FROM EMPLOYEES
;
"""

# Excute the query using pandasql
result = sqldf(query, env={'EMPLOYEES':df})

# Display the result dataframe
display(result)
```
